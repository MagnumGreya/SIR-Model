```python
from matplotlib import legend
import matplotlib.pyplot as plt
import numpy as np

ndays = 100
dt = .1 # time step in days
npts = int(ndays/dt)
beta = (1/3) # contacts per infected per day i.e infection rate
gamma = (1/14) # recovery per infected per day i.e recovery rate

S = np.zeros(npts) # susceptible population
I = np.zeros(npts) # infective population
R = np.zeros(npts) # removed population
t = np.arange(npts)*dt

I[0] = 0.001 #intitial infective fraction of population
S[0] = 1. - I[0] # initial susceptible population
R[0] = 0. # initial recovered population

for i in range(npts - 1):
  S[i + 1] = S[i] - (beta*S[i]*I[i])*dt
  I[i + 1] = I[i] + (beta*S[i]*I[i]-gamma*I[i])*dt
  R[i + 1] = R[i] + (gamma*I[i])*dt

  fig = plt.figure(1); fig.clf();
  plt.plot(t,S,'r', lw = 3,label = 'Susceptible')
  plt.plot(t,I,'g', lw = 3, label = 'Infective')
  plt.plot(t,R,'b', lw=3, label = 'Removed')
  fig.legend(); plt.xlabel('Days'); plt.ylabel('Fraction of population')

```
